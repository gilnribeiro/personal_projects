```python
import os
import pandas as pd
```

```python
cwd = os.getcwd()
working_file = os.path.join(cwd, 'movies-box-office-dataset.csv')
```

```python
df = pd.read_csv(working_file)
```

```python
df
```

```python
df['Rank'] = -1
```

```python
df.head()
```

```python
df.rename(columns={'%':'Domestic %', '%.1':'Foreign %'}, inplace=True)
```

```python
df.head()
```

```python
cols_to_clean = ['Worldwide','Domestic','Foreign']

def currency_str_to_int(current_val):
    currency_val = current_val.replace('$','').replace(',','')
    try:
        currency_val = int(currency_val)
    except:
        # takes any row value not Int and converts to 0
        currency_val = int(0)
    return currency_val

def clean_col(row):
    for col in cols_to_clean:
        current_val = row[col]
        row[col] = currency_str_to_int(current_val)
    return row

df_cleaned = df.apply(clean_col, axis=1)
df_cleaned.head()
```

```python
df_cleaned.dtypes
```

```python
df_cleaned.sort_values(by='Worldwide', inplace=True, ascending=False)
df_cleaned.head()
```

```python
df_cleaned.reset_index(inplace=True, drop=True)
df_cleaned.head()
```

```python
df_cleaned['Rank'] = df_cleaned.index + 1
df_cleaned.head()
```

```python
df_cleaned['Domestic %'] = df_cleaned['Domestic'] / df_cleaned['Worldwide']
df_cleaned['Foreign %'] = df_cleaned['Foreign'] / df_cleaned['Worldwide']
```

```python
df_cleaned.dtypes
```

```python
df_cleaned.to_csv(os.path.join(cwd, 'movies-box-office-dataset-cleaned.csv'), index=False)
```
