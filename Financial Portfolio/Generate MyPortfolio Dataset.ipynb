```python
import datetime as dt
import matplotlib.pyplot as plt
from pandas_datareader import data as web
import pandas as pd
```

```python
# Import and clean data
tickers = ['NIO', 'WKHS', 'CRSR', 'IWDA.AS']
stock_names = ['NIO', 'WorkHorse', 'Corsair', 'iShares World']
shares = [6, 8, 8, 19]
prices = []
total = []

for ticker in tickers:
    df = web.DataReader(ticker, 'yahoo', dt.datetime(2019,8,10), dt.datetime.now())
    price = df[-1:]['Close'][0]
    prices.append(price)
    index = prices.index(price)
    total.append(price * shares[index])

stocks_tickers = ['NIO', 'WKHS', 'CRSR', 'IWDA.AS', 'SXR8.DE']
stocks = web.DataReader(stocks_tickers, 'yahoo', dt.datetime(2019,8,10), dt.datetime.now())
    
total_str = ['$'+str(round(t,2)) for t in total]

stock_names_and_total = stock_names.copy()
stock_names_and_total.append('Total')
total_portfolio = total.copy()
total_portfolio.append(sum(total))
total_portfolio_str = ['$'+str(round(t,2)) for t in total_portfolio]
```

```python
stocks.drop(columns=['Adj Close','High','Low','Open','Volume'], inplace=True)
```

```python
stocks_close = stocks.Close.reset_index().copy()
stocks_close.rename_axis(None, axis = 1, inplace=True)
stocks_close
```

```python
stocks_close.set_index('Date')
stocks_close['My Portfolio'] = (stocks_close['NIO']*shares[0]+stocks_close['WKHS']*shares[1]+stocks_close['CRSR']*shares[2]+
                                stocks_close['IWDA.AS']*shares[3])

stocks_close_dropna = stocks_close.dropna().copy()

stocks_close['Portfolio Return'] = stocks_close['My Portfolio'] / stocks_close['My Portfolio'].shift(1) - 1
stocks_close['Market Return'] = stocks_close['SXR8.DE'] / stocks_close['SXR8.DE'].shift(1) - 1
stocks_close
```

```python
# Export Stocks close with NaN data points
stocks_close.to_csv('stocks_close.csv', index=False)
```

```python
stocks_close_dropna['Portfolio Return'] = stocks_close_dropna['My Portfolio'] / stocks_close_dropna['My Portfolio'].shift(1) - 1
stocks_close_dropna['Market Return'] = stocks_close_dropna['SXR8.DE'] / stocks_close_dropna['SXR8.DE'].shift(1) - 1
stocks_close_dropna
```

```python
# Export this data set to csv
stocks_close_dropna.to_csv('stocks_close_dropna.csv', index=False)
```

```python
stocks_close_dropna['month'] = stocks_close_dropna['Date'].dt.month
stocks_close_dropna['year'] = stocks_close_dropna['Date'].dt.year
stocks_close_dropna.groupby(['year','month'])[['Portfolio Return','Market Return']].sum()
```

```python
stocks_close_dropna.set_index('Date', inplace=True)
stocks_close_dropna
```

```python
group_st_close_dna = stocks_close_dropna.groupby(pd.Grouper(freq="M"))[['Portfolio Return','Market Return']].sum().reset_index()
group_st_close_dna
```

```python
group_st_close_dna.to_csv('grouped_stock_close_dna.csv', index=False)
```
